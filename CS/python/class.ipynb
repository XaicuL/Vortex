## class 사용 방법

위에서의 스토리에서 이제 class를 생각해보자.

- class : 주스 기계
- class instance : 주스 기계로 뽑아낸 주스

사실 여기선 그렇게 많이 구구절절 설명해봤자 코드가 더 중요하다. <br>
코드로 가보자.
