```python
for i in range(1,10): #i 는 loop variable이라는건데, 여기서 컴퓨터 내부에서 일어난 일을 유치하지만 써보자면
#i range를 향해 가 ! (for) 1부터 9까지 가 ! 
    print(i) #대신 하나를 갈 때마다 모두 출력해줘 !

#여기서 이상함을 느껴야 한다.
#왜 9까지야? 10까지 아니야? 라고 할 수 있는데 10진법과 메모리를 배우면 쉽게 설명이 가능하지만
#아직 우리는 해당 내용을 배우지 않았으니 이해를 돕기 위한 설명을 하자면, 
#아까 An에서 우리는 '임의로' 0번째부터 시작하기로 했다 왜냐면 카톡으로 자꾸 5번째 뭐냐 하면 엑셀에서 5. 쳐서 찾고 막 그래야 하는 그런 이상한 알바니까
#근데 10번째 뭐냐 하면 손가락을 두번이나 1 0 해야하니까 얼마나 귀찮았나.. 그래서 0번째부터 할거야 ! 라고 했는데

#그러면 여기서 모순이 생길 수 있다. 0번째부터 하면 우리가 가진 숫자가 만약 10개였다면 0번부터 10번까지면 11개다.
#그러면 여기서 뭔가 이상하다.. 근데 해결 방법은 간단하다. 10번을 9번이라고 하면 된다 ! 즉 마지막 - 1을 해주면 그냥 똑같이 10개가 된다
#그리고 우리는 편하게 0 ~ 9를 물어볼때마다 ㅇㅋ 하면서 엑셀에서 9 , 8 , 5 , 이렇게만 치면 아까의 An이 단순해진다 

#일반화를 해보자.
#for loop 에서 range는 다음과 같은 조건을 따른다
#(start, stop - 1, step)
#이 조건은 항상 기억하자. 단 step은 쓰지 않아도 된다. 그러면 just 1이 된다.

#아래의 코드는 이제 아까 우리가 보고 싶었던 An을 한번 정리해서 가져와봤다.

print('------------------------------------')
for a in range(3,16,3):
    print(a)

#근데 여기서 만약 아 나는 A1 A2 이런 것도 표기가 되면 좋겠어 라고 한다면

print('---------------')
for n in range(1, 6): 
    b = n * 3   
    print(f'A{n} = {b}')  

#근데 또 갑자기 다 더하고 싶을 수 있다
#그러면 다음과 같이 작성하면 된다

total = 0  # 합계를 저장할 variable 

for n in range(1, 6):
    b = n * 3  # n이 1일 때 3, 2일 때 6... 이 되는 규칙
    print(f'A{n} = {b}')
    total += b  # b 값을 total에 계속 누적

print(f'총합 = {total}')

#제발 코드를 그냥 보고 당연한거 아닌가? 이러면 안된다
#하나 하나 뜯어보자.

#total = 0 : Init variable 
#이걸 하는 이유는 단순하다. 일단 total 이라는 variable을 assign 해두는거다.

#for n in range(1, 6):
#n 이라는 loop variable이 1부터 5까지 갈건데 step은 따로 없다.
# b = n * 3
#b 라는 variable은 n 이라는 variable이 1부터 5까지 가는 내내 n에 1,2,3,4,5 가 assign 된다
#그리고 그걸 바로 total 이라는 variable에게 준다 

#여기서 뇌가 꼬일 수 있다. 쉽게 생각하자. 공장이라고 생각하자. 한줄 한줄이 그냥 그 코드가 할 일이다.

#total = 0 :기계 Turn On 정도로 이해하자. 그냥 전원만 켜둔거고 아무것도 안했다 아직

#for n in range(1,6) : 기계를 설정하는건데, 불쌍하다. 기계가 하나 뿐이다. n 이라는 기계 하나로 다 해야한다.
#b = n * 3 : 가장 바쁘다 n이라는 기계가 1번 생산공정에 투입되면 그게 쌓이고 나중에 하면 골치 아프니 바로 n에게 받아온다 
#그리고 그걸 3과 곱해서 자기에게 assign 해둔다.
#total += b : assign 해두고 쉬지 않는다. 바로 달려가야한다 total 라는 기계에 직접 가서 자기가 적어온 보고서를 total 기계에 수기로 기입한다.
#print(f'총합' = {total}) : 이제 산출물이 나온다. 다 출력이 되면 공장이 할 일은 끝이다.

print('-------------------')
#for loop부턴 매우 중요하기에 학교 예제도 가져와본다.

word = "Python" #variable str assign 
for ch in word: #ch 라는 loop variable 이 word 라는 variable에 간다. 
    print(ch) #컴퓨터는 무조건 P 도 y 도 다 다른거로 인식하여 순회한다.

#이제 pass 개념을 도입해보겠다.

for i in range(1, 20): #여기선 분명 19까지 출력이 되야 정상이다.
    if i < 15: #만약 i가 15보다 작다면 ? 
        pass #pass 해 !
        print(i) #그리고 그걸 출력해 ! 
#그럼 여기서 이제 왜 14까지만 출력된건지 이해 되었으리라 믿는다
#15보다 커지면? 이라는 조건이 없고, 그때 뭘 하라는지 컴퓨터는 모른다. 단지 탈출하고 출력하라는 것만 알고 있다.

print('-------------------------------while loop----------------------------------------')

i = 1
n = 10
total = 0

while i < n:
    total += i
    i += 1

print(f'Sum of i : {total}')


#i = 1 , n = 10, total = 0 이런게 왜 필요하지? 의문이 들 수 있다.
#그러면 좋은 코드를 소개 하겠다.

# while i < 100:
#     print(i)
#해당 코드를 실행하면 어디선가 비행기 이륙소리가 날거다. 
#그럼 안전 장치는 뭐가 없을까 ? 

# 밑에 이걸 추가하면 된다
# i += 1

#그리고 함부로 while에선 조건부에서 상수를 쓰지 않는 이유 또한 안전 장치라고 생각하면 매우 좋다.
# break , continue 같은 좋은 function이 있으나, 그건 뒤에서 다루기로 한다.
```
