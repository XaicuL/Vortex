## 2) Tuple : Not A But B ~= 순서는 있으나 변할 수 없는

**Tuple** 은 기본적으로 앞서 배운 list와 공통적인 부분이 있으나, 공통적이지 않은 부분도 함께 존재한다.
<br> **list** 는 분명 순서도 있고, 변할 수 있는(mutable) 이 특징이었고 가장 중요한데,
<br> **tuple**는 순서는 존재한다 **BUT**  tuple은 **immutable** 이다. 변할 수 없다.

<br> 그래서 내용도 단순 할 수 있지만, 여기서 내가 바라는 점은 **list와 차이점은 뭐지??** 라는 점을 계속 질문을 던지면 좋겠다.

## 3) Dictionary : Dict는 변할 수 없는 초판과 같다.

**dictionary**는 굳이 굳이 이걸 어떻게 이해해야하고 저렇고를 말을 하지 않아도 될 것 같다.
<br> 그저 **사전** 이다 우리가 사전을 넘기다 Apple 라는 단어를 발견했다. 그럼 뭐라고 써있지? 사과라고 써있다.
<br> 그 어떤 사전에도 person. 만 놓고 이후에 뭔가 대응되는 개념이나 명사를 써주지 않는 사전은 없다.
<br> dict는 정의는 다음과 같다.

**Def** : 키(Key)와 값(Value)을 한 쌍으로 저장하는 자료형
<br> 개인적으로 나는 다음과 같이 이해한다.
<br> (내가 산게 영어 - 한글 사전이라 가정한다,)

- Key : 영어
Ex ) Apple

- Value : 한글
Ex ) 사과

<br>
그럼 내가 왜 **변할 수 없는 초판** 과 같다고 했을까?
<br> 그건 이유가 있는데 바로 **key** 는 **immutable** 인 성격을 가진다.
<br> 다만 **Value** 는 **mutable** 이다 언제든 변할 수 있다.

<br> Ex)
- Key : Trot

- Old Value : 빨리 걷다, 속보.
- New Value : 트로트. (너가 아는 그 트로트 맞다.)

<br> 이렇듯 Key는 **fix**가 되어있어야 한다.
<br> 그런데 Valdue는 **un fix.** 즉 변할 수 있다.

```python
a = {1,2,3,3}
print(f'a = {a}')

#여기서 어??????? 하면 좋겠다.
#분명 a 라는 set 안에는 3이 2개나 포함되어 있다.
#그런데 출력을 보면 3이라는 element가 하나로 인식된다.

#이게 바로 set만이 가진 중복 불가 라는 성질 때문이다.
#이유는 무엇일까? 이유는 정말 수학적으로 생각하면 단순한데
#수학에서의 집합 정의를 보면 '서로 다른 원소들의 모임' 이다.

#즉 수학적으로 서로 '다른' 이라고 했으니 python에서의 set 또한 서로 다르게 해야한다는 원칙을 세워둔 것
#이제 이걸 좀 괴랄하게 표현해서 보여주겠다.

b = {1,1,1,1,1,1,1,1,11,1,11,1,1,1,1,1,4,43,24,2343,22,2,2,2,2,22,2,2,4,4,4,4,44,4,44,4,4,4,2,2,22,324,24,234,234}
print(f'b = {b}') 
#이래서 우리가 set을 알아야 한다.
#현실에서의 data나 무언가 트래픽이 쏟아져내리더라도, 우린 거기서 기준을 설정해야한다
#그런데 현실에서의 data는 중복이.. 정말 많다... 예를 들어 이런 상황을 생각해보자.
#우리가 매출을 좀 이상하게 정리해야한다. 
#어떤 날엔 담배가 엄청 많이 팔렸는지 4500원이 1000개가 넘는다.
#그리고 어떤 날엔 새콤달콤이 너무 많이 팔린건지 500원라는 data가 20000개가 넘는다 
#근데 우리는 .. 일단 엑셀에 4500원, 1000원, 500원 이걸 열로서 기입해두고 싶다
# 하.. 그럼 set을 쓰지 않으면 하루종일 1000원 아 이미 썻어 2000원 2000원 아 2000원 방금.. 2100원 2130원 2145원... 2000원 아 2000원 아까 ;;;;
# 근데 set을 쓰면 딸깍 한번이면 바로 중복을 제거해준다
# 이러한 점에서 우리는 set의 중복 제거에 감탄을 할 수 밖에 없다.

#이제 set에서의 연산을 보자.

print('---합집합---')

set_1 = {1,2,3,4,5,6}
set_2 = {5,6,7,8}

sum_sol1 = set_1 | set_2 
sum_sol2 = set_1.union(set_2) #union은 list로 치면 Append 느낌이다. method고, | 는 연산자다.

print(f'합집합_1 {sum_sol1}')
print(f'합집합_2 {sum_sol2}')

#여기서 엥...? 할 수 있는데 합이라는걸 기억하자.
#분명 우리는 set_1 과 set_2 를 합친거다. 근데 set_1의 element만 해도 잠시만
len_set_1 = len(set_1)
print(len_set_1)
#len이 6이라고 한다
#근데 아 잠만

len_set_2 = len(set_2)
print(len_set_2)

# set_1은 length가 6 그리고 set_2 는 4인데

len_sum = len(sum_sol1)
print(len_sum)
#어라 왜 len이 8이 나오지 ?

#그 이유는 합 뒤에 붙은 말인 '집합' 이라는 말을 기억하자.
#결국 합집합 또한 중복은 제거한다. 합집합은 중복은 포함시키지 아니하고 서로만 가지고 있는 element만 더해서 보여준다
#중간에 내가 검증하는 코드를 작성했는데, 부디 여기까지 읽고 있는 누군가라면 python을 이렇게 그냥 내가 해야하는 일을 대신 해주는 언어라고 생각하면 좋겠다

#그러면 여기서 궁금한건 이걸 조건문으로 검증하면 T일까 F일까? 확인해보자 틀리면 뭐.. 그렇구나 하고 지식 +1 이고 맞다고 하면 아하 그렇구나 하고 지식 +1이니 언제든 써먹으면 된다


if len_set_1 + len_set_2 == len_sum:
    print('True')
else:
    print('False')

#사실 간단하다 ㅋㅋㅋ len은 단지 길이니까 아까 6, 4, 8 나온걸.. 6+4 == 8 해보면 되는데 
#한번이라도 더 조건문을 ... ㅎㅎ;;

print('---교집합---')

set_1 = {1,2,3,4,5,6}
set_2 = {5,6,7,8}


intersec_sol1 = set_1 & set_2 
intersec_sol2 = set_1.intersection(set_2) #intersection은 list로 치면 Append 느낌이다. method고, & 는 연산자다.

print(f'교집합_1 {intersec_sol1}')
print(f'교집합_2 {intersec_sol2}')

#교집합은 두 집합에만 존재하는 element를 의미한다.

print('---차집합---')

set_1 = {1,2,3,4,5,6}
set_2 = {5,6,7,8}


diff_sol1 = set_1 - set_2 
diff_sol2 = set_1.difference(set_2) #difference은 list로 치면 Append 느낌이다. method고, - 는 연산자다.

print(f'차집합_1 {diff_sol1}')
print(f'차집합_2 {diff_sol2}') 

#차집합은 set_1에는 속하지만, set_2에는 속하지 않는 element를 추출하는데 그럼 반대로 하면 어떻게 될까?
#한번 해보자

diff_sol3 = set_2 - set_1

print(f'차집합 set_2 - set_1 = {diff_sol3}') #반대로 set_2에는 속하지만 set_1에는 속하지 않는 element가 추출된다.

#사실 더 많은걸 다룰 수 있다. 다만 고교 과정에서 배운 수학으로는 여기서 일단은 멈추고 method를 소개하고자 한다.

print('----------------------------------------------------')

#element를 추가하기 위해선 add(element) 를 사용하면 되는데
#여기서 중요한 점을 한번 다루고 가자. 일단 코드부터 보자

x = {1,2,3,4,5,6,7}
print(f'current set x = {x}')

x.add(8)
print(f'New set x = {x}')

#8이라는 element가 추가 됨을 알 수 있다.
#그러면 만약 이미 있는걸 추가하면 어떻게 될까? 해보자.

x.add(3)
print(f'New set x (2) = {x}') #그렇다. 중복은 불가능하다는 1원칙이 우선이다.

#그러면 우리의 뇌의 온도를 올려주었던. Return과 method 를 다시 다뤄보자.

y = x.add(9)
print(y) #여기서도 None이 나온다 

#왜 None인지 이제 이해를 했겠지만 너무너무 중요한 개념이라서 한번 더 짚고 넘어간다

#set_name.add(element)의 우선순위는 다음과 같다 
# 1. set_name이라는 set에 element를 배송해준다
# 2. 다시 물류센터로 온다

#근데 여기서 고객이 직접 와서는 "제 택배 주세요."  그러는거다
#이미 근데 우리는 배송을 완료했다. 이미 해당 고객 집 문 앞에 두고 왔다.
#그 고객이 바로 좌변 즉 y가 되는거다. 

#y라는 고객 입장에서도 사실 타당하다.
#고객은 결국 배송물품을 받아야 한다.
#그래서 물류센터로 왔는데 이미 배송 기사님께서 배송을 집 문 앞에 배송을 하고 오셨다
#그럼 하나를 더 받을 수 있을까? 이미 집 앞에 있는데? 그건 불가능하다.
#그래서 None을 assign 받는 것이다. 택배사 입장에선 당장 줄 것이 없으니까

#element를 2개 이상 추가를 해야할 땐 update([element1, element2]) 로 하면 된다 아래의 코드를 보자.

x.update([11,12])
print(f'New element (3) : {x}') #정상적으로 배송이 되었음을 확인했다. return none에 대해선... 일단 이제 넘어가도 될 것 같다.

#element 삭제는 어렵지 않다 
#remove(element) 더군다나 중복도 안되서 더 직관적이다.

x.remove(12)
print(f'New element (4) : {x}')
```
