## 1) Output : print의 여러가지 방법
기본적으로 python에 출력을 담당하는 function은 **print** 인데, <br>
print function 내부에서도 정말 많은 출력 방법이 존재한다. <br>

- print(x) : x라는 variable을 출력하는 방법이다. <br>추후 type을 다룰 때 더 설명하겠지만 위에 출력 방식과 엮어서 사용하는 편이다.

- print(f'') : 가장 최근에 나온 출력 방법이다. 가장 간단하다. 개인적으로 매우 추천하는 방식이다. 

```python
print('Helloworld')

#다음을 출력하면 기본적으로 출력이 정상적으로 된다. 
#나머지 두 방법은 variable을 소개하고, 더 자세히 다뤄보도록 하자.
```

## 2) variable : variable = Assign 그리고..

**변수** 란 수학적인 개념에선 변할 수 있는 **수** 라고 정의를 한다. <br>
다만 이건 엮여있는게 많지만 ~~결국 python도 이진수로 표현하고 인터프리터를 거쳐~~ 미안하다. <br>
변수는 결국 언제든 우리가 마음대로 변하게 할 수 있는 값이락 이해를 하고 넘어가자. 
<br>

다만 이건 양보를 못하겠다. **변수**는 기본적으로 **Assign 한다** 라고 말을 한다 <br>
한글로 직역하면 **변수 할당** 인데, 본인은 Assign 이라고 하는 편이다.
<br>

Assign의 방법은 어렵지 않다. 단순히 자신이 변수 이름을 네이밍 하고 **=** 을 사용하면 된다. <br>
Ex)<br>

x = 5
<br>

이렇게 하면 x 라는 variable에는 5 라는 값이 Assign 된 것이다.
<br>

그런데 내가 제목에서 '그리고..' 라는 말을 했는데 그 이유는 매우 단순하다. <br>
**changing binding** 이라는 개념을 소개해주고 싶었기 때문인데 <br>

changing binding 이란 너무 깊게 들어가면 다른 언어의 영역이니 python 선에서 마무리 해보자면 <br>
다음과 같은 상황을 말한다. 아래 예시를 보자. <br>

Ex)  <br>

x = 3.14

<br>
y = x + 4

x = x - 0.14

print(x)

<br>

이때 출력이 되는 값은 ? ~~3.0이 출력되야 정상이다.~~ <br>

이렇듯, variable의 이름을 중복으로 쓰고 새로운 값을 binding 하면 <br>
Old variable 즉 초기 변수의 값은 사라지고, New variable의 값이 새로운 변수의 값이 된다 <br>

이제 밑에 코드에서 print도 더 다뤄보며 깊게 가보자.

```python
x = 1

print('1. x = ',x) 
#' ' 안에 들어가면 print는 그냥 그대로 출력을 해버린다. 
#하지만 variable은 그렇게 출력이 되어선 안된다 variable은 가면을 쓴 상태지 그 가면이 진짜 얼굴이 아니다.
#다음과 같이 작성을 하면 x = 1 이라고 출력은 된다.
#Cuz, '' 다음 쉼표(,) 를 찍고 따옴표를 쓰지 않은 상태로 variable call을 했기 때문인데 
#그러면 x = 1 이라고 출력이 될 것 이다.

print('2. x = x')
#이 예시가 더 극적일 것 같다. 출력하면 분명 x = x 라고 무미건조하게 나올 것이다.
# 쉼표 있고 없고 그리고 따옴표의 차이가 엄청난 결과의 차이를 만들어낸다

print(f'3. x = {x}')
#f - string 이라 부르고 f - str 이라고 부른다
#이 방식을 사용하는 방법은 따옴표 앞에 f를 쓰고 따옴표를 쓰면 된다
#그리고 variable을 call할 땐 {} 를 쓰고 그 안에 variable을 call 하면 된다

print('-------------------------------------------------------')

x = 1 
y = 3

print('1. x + y = ' ,x+y)
print('2. x + y = x + y ')
print(f'3. x + y = {x+y}')


print('-------------------------------------------------------')

#changing binding을 다루기 전에 사칙 연산부터 다루고 진행

a = 4
b = 2

plus = a + b
min = a - b
sq = a * b
div = a / b #개인적으로 생각하는 개념으로는 분수라고 생각하는 편이다. 그렇게 보면 a 는 분자 b 는 분모가 된다.

#이제 f-str을 적극적으로 이용해보자.

print(f'a + b = {plus}')
print(f'a - b = {min}')
print(f'a x b = {sq}')
print(f'a / b = {div}')


print('-------------------------------------------------------')

# changing binding 

m = 1
n = 2

print(f'current m = {m}')

plus2 = m + n


print(f'm + n = {plus2}')

m = 4  #changing binding 

print(f'New m = {m}')
```

## 3) input : input의 여러가지 쓰임

print function은 기본적으로 output을 담당한다. <br>
여기서 **out** 이라는 말의 어감을 살린다면 **안에서 밖으로** 라고 할 수 있다. <br>

그러나 **input** function은 밖에서 안으로 넣는 것인데. <br>
여기서 밖을 우리 클라이언트 (쉽게 코드치는 우리라고 지금은 생각하자.) 클라이언트를 out이라고 본다면 <br>
in 은 IDE(쉽게 말해 지금 우리의 터미널 혹은 코딩을 하는 프로그램이라고 생각하자.)로 불러온다라고 생각하면 어감을 살려낼 수 있다 <br>

**=>** input으로 variable을 assign 해두면 실행을 할 시 우리에게 값을 입력하라고 한다. <br>
우리가 입력하기 전까진 해당 variable은 assign이 되지 않은 상태다. <br>
여기서 **variable**의 성격이 나온다. 언제든 **변 할 수 있 는** 이니까. <br>

**다만** 주의해야 할 지점은 **input function** 은 기본적으로 **str** 이라는 type으로 variable을 assign 한다는 것이다. <br>

### TYPE

- int : 정수형, 간단히 소수점이 없는 숫자라고 생각하면 편할 것 같다.

- float : 부동소수점, 간단히 소수점이 존재하는 숫자라고 생각하면 편할 것 같다.

- str : 문자열, 간단히 문자(언어) 라고 생각하자. 
cf) 문자열을 출력할 땐 무조건 '(따옴표)' 를 기입해야 했었다. 그런데 variable을 해당사항이 아니었고 <br>
int 및 float 또한 따옴표를 기입하지 않아도 된다.

### type() function

**Def** : 쉽게 말하면 현재 해당 변수의 type을 안내해주는 funcion이다. <br>

이제 아래의 예제 코드들에서 한번 간단히 다루어 보겠다.
다만 나는 모든 입력값을 1 로 두고 진행한다는 점을 미리 고지한다.


```python
x = input('x = ')
print(f'x = {x}') #원한다면, 자신이 원하는 숫자를 입력해보아도 좋다.

print(f'x type is {type(x)}')

print(f'입력한 x = {x}')

print('-------------------------------------------------------')

#지금부턴 type changing 을 해볼건데, 이건 input function에서 매우 좋은 기능이니 
#부디 잘 이해하면 좋겠다. 
#사용법은 아래 코드와 같다

int_x = int(input('int x = ')) #str -> int , 입력은 3으로 하면 좋겠다.
float_x = float(input('float x = ')) #str -> float, 입력은 3.14 해주면 좋겠다.

print(f'int_x = {int_x}')
print(f'float_x = {float_x}')

print(f'int_x type is {type(int_x)}')
print(f'float_x type is {type(float_x)}')

print(f'입력한 int_x = {int_x}')
print(f'입력한 float_x = {float_x}')
#cf) input으로 assign 된 variable로 사칙연산 code를 직접 짜보면 좋겠다.
#덧셈만 예시로 남겨주면..


a = int(input('a = '))
b = int(input('b = '))

print(f'입력한 a = {a}')
print(f'입력한 b = {b}')

print(f'a + b = {a+b}')
#OR
plus = a + b
print(f'a + b = {plus}')
```
